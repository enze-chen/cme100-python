## 2. Plotting

This is done with the powerful [Matplotlib](https://matplotlib.org/) library.

### (a) Basic plotting

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 15, 0.1)
y = np.sin(x) * x

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel="Time", ylabel="Position", title="Pretty plot")
plt.show()
````

We can easily plot multiple datasets on a single set of axes.

````python
z = np.cos(x) * x ** 0.5

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, z, ls='--')
ax.set(xlabel="Time", ylabel="Position", title="Pretty plots")
plt.show()
````

Or we can put them on separate axes on the same figure.

````python
fig, ax = plt.subplots(figsize=(10,8), nrows=2, ncols=2)
ax[0,0].plot(x, y)
ax[0,1].plot(x, z, ls='--')
ax[1,0].plot(y, z, ls=':')
plt.show()
````

## 3. Plotting in 3D

This is still done in Matplotlib, but with some special function parameters.

### (a) Parametric curves

````python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 4*np.pi, 0.1)
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.plot(x, y, z)
plt.show()
````

### (b) Surfaces

````python
import numpy as np
X, Y = np.meshgrid(np.arange(-2, 2, 0.1), np.arange(-2, 2, 0.1))
Z = np.sin(X ** 2 + Y ** 2)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma', antialiased=False)
plt.show()
````

## 7. Programming in Python

As with most problems, there are _multiple_ possible solutions, and the idea is we pick the one that is most logical and/or efficient.

````python
def factorial_for(n):
    product = 1
    for i in range(2, n+1):
        product *= i
    return product

def factorial_while(n):
    product = 1
    i = 1
    while i <= n:
        product *= i
        i += 1
    return product

def factorial_if(n):
    product = 1
    i = 1
    while True:
        product *= i
        i += 1
        if i > n:
            break
    return product

print(factorial_for(8), factorial_while(8), factorial_if(8))
````
