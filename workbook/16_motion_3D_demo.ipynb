# Motion of a particle

We will continue working with 3D plots, this time to visualize the 3D motion of a particle.

## Summary of commands

In this exercise, we will demonstrate the following:
- [`np.gradient(y, x)`](https://numpy.org/doc/stable/reference/generated/numpy.gradient.html) - Computes the numerical derivative of a function `y` with a spacing defined by `x`.
- [`np.arccos(x)`](https://numpy.org/doc/stable/reference/generated/numpy.arccos.html) - Computes the inverse cosine of `x`.

The motion of a particle in three dimensions is given by the following parametric equations:

$$ 
\begin{align*}
x(t) &= \cos(t) \\
y(t) &= 2\sin(t) \qquad \text{for} \quad 0 \le t \le 2\pi \\
z(t) &= 0
\end{align*}
$$

## Part (a)

Let $\Delta t = 0.1$. 
Plot the particle's trajectory in 3D using [`ax.plot()`](https://matplotlib.org/stable/api/_as_gen/mpl_toolkits.mplot3d.axes3d.Axes3D.plot.html).

```python
import numpy as np
import matplotlib.pyplot as plt

# values
dt = 0.1
t = np.arange(0, 2*np.pi, dt)
x = np.cos(t)
y = 2 * np.sin(t)
z = 0 * t   # so it's the same length as the other arrays

# make a plot
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.plot(x, y, z)
ax.set(xlabel='x', ylabel='y', zlabel='z')
plt.show()
```

## Part (b) 

Compute the velocity components $v_x$, $v_y$ and the magnitude of the velocity vector.

```python
vx = np.gradient(x, t)
vy = np.gradient(y, t)
speed = np.sqrt(vx**2 + vy**2)
speed
```

## Part (c)

Compute the acceleration components $a_x$, $a_y$ and the magnitude of the acceleration vector.

```python
ax = np.gradient(vx, t)
ay = np.gradient(vy, t)
accel = np.sqrt(ax**2 + ay**2)
```

## Part (d)

Compute the components of the unit tangent vector $\vec{T} = \begin{bmatrix} T_x & T_y & T_z \end{bmatrix}$.

```python
Tx = vx / speed
Ty = vy / speed
```

## Part (e)

Compute the angle $\theta$ between the velocity vector and the acceleration vector.

```python
theta = np.arccos((vx * ax + vy * ay) / speed / accel)
theta
```
